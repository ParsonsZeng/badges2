## Statistics & p-values calculation

```python
attention_horizon = 14


def get_pvalues(events):
    times = np.array(events["start_time"]+events["action"])

    st = events["start_time"][0]
    et = events["max_time"][0]

    b1 = max(events.get("tag_editor", [-INF])[0], TAGEDITOR_INTRO)
    b2 = max(events.get("research_assistant", [-INF])[0], RESEARCHASSISTANT_INTRO)
    
    if not b2<INF: return None, None

    bp, llr_true, ll1, ll2, l0, l1, l2 = bootstrap_pvalue(max(b1, st), et, events["action"], b2, 
                                                          attention_horizon=attention_horizon)
    tp = 1-chi2.cdf(2*(llr_true), 1) if llr_true is not None else None
    
    return bp, tp, l1, l2, llr_true
```

```python
def plot_user(uid):
    events = id2events[uid]
    actions = events.get("action", [])
    plot_events(events)
    bp, tp, l1, l2, llr_true = get_pvalues(events)
    pyplot.title("s=%.0f e=%.0f f=%.0f l=%.0f a=%.0f bp=%g tp=%g llr=%g" % 
                (events["start_time"][0], events["max_time"][0], 
                 min(actions), max(actions), 
                 len(actions), bp, tp, llr_true))
    
    plot_badge(events, "tag_editor")
    plot_badge(events, "research_assistant")
    plot_badge(events, "research_assistant*")
    
    plot_badge_intro(events, "TAGEDITOR", TAGEDITOR_INTRO, color="blue")
    plot_badge_intro(events, "TAGEDITORM", TAGEDITOR_MOD, color="blue")
    plot_badge_intro(events, "RESEARCHASSISTANT", RESEARCHASSISTANT_INTRO, color="blue")
    
```

```python
def plot_user2(uid):
    events = id2events[uid]
    actions = np.asarray(sorted(events["action"]))
    
    pyplot.scatter(actions, np.zeros(len(actions)), marker="x")
    pyplot.axvline(events["research_assistant*"])
    pyplot.xlim((events["research_assistant*"][0]-1, events["research_assistant*"][0]+1))

    
#plot_user2(1288)
```

```python
#count actions in intervals, intervals and bootstrap p-value bp
data = []
for progress, (i, events) in enumerate(id2events.items()):
    if progress%50000==0: print("%i/%i" % (progress, len(id2events)))
    if len(events.get("action", []))<=0: continue
    times = np.array(events["start_time"]+events["action"])

    st = events["start_time"][0]
    et = events["max_time"][0]

    b1 = max(events.get("tag_editor", [INF])[0], TAGEDITOR_INTRO)
    b2 = max(events.get("research_assistant", [INF])[0], RESEARCHASSISTANT_INTRO)
    
    if not b2<INF: continue

    na1 = sum( (times>st) & (times<=b1) )
    na2 = sum( (times>b1) & (times<=b2) )
    na3 = sum( (times>b2) & (times<=et) )
    
    bp, tp, l1, l2, llr = get_pvalues(events)
        
    data.append( (i, len(events["action"]), 
                  st, b1, b2, 
                  na1, max(b1, st)-st, 
                  na2, min(b2,et)-b1,  
                  na3, et-b2,
                  l1, l2,
                  bp, tp, llr) )
```

```python
COLNAMES = ["id", "actions", 
            "st", "b1", "b2", 
            "na1", "t1", "na2", 
            "t2", "na3", "t3", 
            "l2", "l3",
            "bp", "tp", "llr" ]
data = pd.DataFrame(data).rename(columns=dict(enumerate(COLNAMES)))
data["b*"] = data["bp"].apply(lambda p: "*" if p<0.05 else "")
data["*"]  = data["tp"].apply(lambda p: "*" if p<0.05 else "")
```

```python
# Load the library with the iris dataset
from sklearn.datasets import load_iris

# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier

# Load pandas
import pandas as pd

# Load numpy
import numpy as np
```

```python
data["class"] = (data["*"]!="").astype(int)
data.head()

features = ["age", "reputation", "views"]
features = data[features]
y = data["class"]
```

```python
features = features.as_matrix()
```

```python
ixs = list(range(len(features)))
np.random.shuffle(ixs)

train_ixs = ixs[0: len(features)//2]
test_ixs = ixs[len(features)//2: len(features)]

train = features[train_ixs]
train_y = y[train_ixs]
test = features[test_ixs]
test_y = y[test_ixs]
```

```python
#impute missing values
from sklearn.preprocessing import Imputer
imp = Imputer(axis=0, copy=True, missing_values='NaN', strategy='mean', verbose=1)
imp = imp.fit(train) #REMEMBER TO FIT ON TRAIN SUBSET ONLY!
train = imp.transform(train)
test = imp.transform(test)
```

```python
pd.DataFrame(train)
```

```python
# Create a random forest classifier. By convention, clf means 'classifier'
clf = RandomForestClassifier(n_jobs=30)


# Train the classifier to take the training features and learn how they relate
# to the training y (the species)
clf.fit(train, train_y)
```

```python
preds = clf.predict(train)
pd.crosstab(train_y, preds, rownames=['Actual'], colnames=['Predicted'])
```

```python
preds = clf.predict(test)
pd.crosstab(test_y, preds, rownames=['Actual'], colnames=['Predicted'])
```

## Survival analysis

```python
u2badgeday = dict((u, np.ceil(events["research_assistant"][0])) for u, events in id2events.items() if "research_assistant" in events)
    
```

```python
id2events[1114]["research_assistant"][0]
```

```python
user2survival = {}
user2finite = {}
for u, events in id2events.items():
    if "research_assistant" in events:
        actions = np.array(events["action"])
        last_before = max(actions[actions<u2badgeday[u]])
        first_after = min(list(actions[actions>u2badgeday[u]])+[INF])
        user2finite[u] = first_after<INF
        first_after = min(first_after, user2max[u])
        survival = first_after-last_before
        print(u, last_before, u2badgeday[u], first_after, survival)
        user2survival[u] = survival
```
