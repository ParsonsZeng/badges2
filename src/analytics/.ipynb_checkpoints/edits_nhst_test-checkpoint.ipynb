```python
def test(ll1, ll0, freedom=1):
    if ll1==0 and ll0==0: return 1.0
    return 1-chi2.cdf(2*(ll1-ll0), freedom)


def llr0(start, end, actions, badge, verbose=False):
    actions = np.asarray(actions)
    
    t0 = (end-start)
    t1 = (badge-start)
    t2 = (end-badge)
    count0 = sum((actions>start) & (actions<end))
    count1 = sum((actions<badge) & (actions>start))
    count2 = sum((actions>badge) & (actions<end))
    
    l0 = count0 / t0
    ll0 = l0*count0 - l0*t0
    
    l1 = count1 / t1
    ll1 = l1*count1 - l1*t1
    
    l2 = count2 / t2
    ll2 = l2*count2 - l2*t2
    
    llr = ll2+ll1 - ll0
    
    if (verbose): 
        print("start=%f end=%f badge=%f" % (start, end, badge))
        print("na0=%i t0=%f na1=%i t1=%f na=%i t2=%f" % (count0, t0, count1, t1, count2, t2))
        print("l0=%g ll0=%g  l1=%g l2=%g ll1=%g llr=%g" % (l0, ll0, l1, l2, ll1+ll2, llr))
    return llr, ll1, ll2, l0, l1, l2


def llr(start, end, actions, badge, verbose=False):
    llr, ll1, ll2, l0, l1, l2 = llr0(start, end, actions, badge, verbose=verbose)
    return llr
    

def bootstrap_pvalue(start, end, actions, badge, n=100, margin_events=1):
    llr_true, ll1, ll2, l0, l1, l2 = llr0(start, end, actions, badge, verbose=True)
    actions = np.asarray(sorted(actions))
    actions = actions[(actions>=start) & (actions<=end)]
    if len(actions)<=margin_events:
        print("WARNING: %i actions available with start=%f end=%f" % (len(actions), start, end))
        return 1-chi2.cdf(2*(llr_true), 1), llr_true, ll1, ll2, l0, l1, l2

    start1, end1 = actions[margin_events-1]+EPS, actions[-margin_events]-EPS
    llr_virtual = [(llr(badge, end, actions, virtual_badge) 
                    if virtual_badge>badge 
                    else llr(start, badge, actions, virtual_badge))
                    for virtual_badge in np.random.uniform(low=start1, high=end1, size=n)]
    #print(llr_true, llr_virtual)
    bpvalue = float(sum(llr_true<np.asarray(llr_virtual))) / len(llr_virtual)
    return bpvalue, llr_true, ll1, ll2, l0, l1, l2
```

```python
def get_pvalues1(events):
    times = np.array(events["start_time"]+events["action"])

    st = events["start_time"][0]
    et = events["max_time"][0]

    b1 = max(events.get("editor", [INF])[0], EDITOR)
    b2 = max(events.get("strunk_&_white", [INF])[0], STRUNK_AND_WHITE)
    b3 = max(events.get("copy_editor", [INF])[0], COPY_EDITOR)
    
    if not b2<INF: return None, None

    start, end, badge = max(b1, st,  events["action"][0]), min(et,b3), b2
    bp, llr_true, ll1, ll2, l0, l1, l2 = bootstrap_pvalue(start, end, events["action"], badge)
    tp = 1-chi2.cdf(2*(llr_true), 1) if llr_true is not None else None
    
    return bp, tp, l1, l2, start, end, badge
```

```python
def get_pvalues2(events):
    times = np.array(events["start_time"]+events["action"])

    st = events["start_time"][0]
    et = events["max_time"][0]

    b1 = max(events.get("editor", [INF])[0], EDITOR)
    b2 = max(events.get("strunk_&_white", [INF])[0], STRUNK_AND_WHITE)
    b3 = max(events.get("copy_editor", [INF])[0], COPY_EDITOR)
    
    if not b2<INF: return None, None

    start, end, badge = b2, et, b3
    bp, llr_true, ll1, ll2, l0, l1, l2 = bootstrap_pvalue(start, end, events["action"], badge)
    tp = 1-chi2.cdf(2*(llr_true), 1) if llr_true is not None else None
    
    return bp, tp, l1, l2, start, end, badge
```

```python
def _extract_position(events, v):
    actions = np.asarray(events["action"])
    pos = sum(v>actions)-1 
    return pos
    
def _plot_badge(events, l, color="black"):
    if l not in events: return
    v = events[l][0]
    pos = _extract_position(events, v)+0.5
    pyplot.axvline(x=pos, color=color, lw=2)
    ymin, ymax = pyplot.gca().get_ylim()
    pyplot.text(pos, ymin, " %s (%.2f)" % (l, v),
                rotation=90, fontsize=8, color=color,
                verticalalignment="bottom", horizontalalignment="right")
    
def _plot_badge_intro(events, l, v, color="blue"):
    pos = _extract_position(events, v)+0.5
    pyplot.axvline(x=pos, color=color, lw=2)
    ymin, ymax = pyplot.gca().get_ylim()
    pyplot.text(pos, ymax, " %s (%.2f)" % (l, v),
                rotation=90, fontsize=8, color=color,
                verticalalignment="top", horizontalalignment="right")    
    
def plot_user(uid):
    events = id2events[uid]
    actions = np.asarray(sorted(events["action"]))
    
    diffs = np.diff(sorted(list(actions)+events["max_time"]))
    xs = list(range(len(actions)+1))
    pyplot.plot(xs, list(diffs)+[diffs[-1]], drawstyle="steps-post", lw=2, color="dodgerblue")
    
    positions = [i for i, t in enumerate(sorted(actions)) if i%10==0]
    labels = ["%i (%.1f)" % (i, t) for i, t in enumerate(sorted(actions)) if i%10==0]
    pyplot.xticks(positions, labels, rotation=45, horizontalalignment="right")#, rotation_mode="anchor")
    
    #delay between start_time and the next action
    pos = _extract_position(events, events["start_time"][0])
    nxt = actions[pos+1]
    delay = nxt - events["start_time"][0]
    pyplot.plot([pos, pos+1], [delay, delay], drawstyle="steps-post", lw=2, color="red")

    bp, tp, l1, l2, start, end, badge = get_pvalues1(events)
    pyplot.title("ss=%.0f ee=%.0f f=%.0f l=%.0f a=%.0f bp=%g tp=%g" % 
                (start, end, 
                 min(actions), max(actions), 
                 len(actions), bp, tp))
    pyplot.yscale("log")
    pyplot.grid(True)
    pyplot.xlabel("action number (time in days)")
    pyplot.ylabel("time to next action (days)")
    pyplot.xlim((-2, len(actions)+1))
    
    for b in badges:
        _plot_badge(events, b)
            
    _plot_badge_intro(events, "EDITOR", EDITOR, color="blue")
    _plot_badge_intro(events, "STRUNK_AND_WHITE", STRUNK_AND_WHITE, color="blue")
    _plot_badge_intro(events, "COPY_EDITOR", COPY_EDITOR, color="blue")
    
```

```python
data = []
for progress, (i, events) in enumerate(id2events.items()):
    if progress%100000==0: print("%i/%i" % (progress, len(id2events)))
    if len(events.get("action", []))<=0: continue    
    if "strunk_&_white" not in events: continue
    
    bp1, tp1, l1, l2, s1, e1, b1 = get_pvalues1(events)
    bp2, tp2, l3, l4, s2, e2, b2 = get_pvalues2(events)
    
    data.append( (i, len(events["action"]), 
                 bp1, tp1, l1, l2, s1, e1, b1,
                 bp2, tp2, l3, l4, s2, e2, b2) )
```
